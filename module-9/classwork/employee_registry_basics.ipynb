### ЗАДАЧА: Реестр сотрудников (база ООП)

Нужно реализовать простой класс сотрудника и класс реестра сотрудников.

НЕОБХОДИМО РЕАЛИЗОВАТЬ:

1. Класс `Employee` с полями `emp_id`, `name`, `salary`, `department`.

2. В `Employee` добавить:
   - метод `raise_salary(percent)`
   - метод `to_dict()`
   - `@property salary` с проверкой (зарплата не может быть отрицательной).

3. Класс `EmployeeRegistry`:
   - хранит сотрудников в словаре `emp_id -> Employee`
   - метод `add_employee(employee)`
   - метод `remove_employee(emp_id)`
   - метод `find_by_department(department)`
   - метод `total_payroll()` (сумма зарплат всех сотрудников).

4. Добавить сотрудников из исходного списка, сделать 1-2 повышения зарплаты и вывести итоговый отчёт.

ПОДСКАЗКИ ПО ЛОГИКЕ:
- `raise_salary(percent)` -> увеличить зарплату: `salary = salary * (1 + percent / 100)`.
- `to_dict()` -> вернуть словарь с ключами: `emp_id`, `name`, `salary`, `department`.
- `find_by_department(department)` -> вернуть список объектов `Employee` только нужного отдела.
- `total_payroll()` -> сумма `employee.salary` по всем сотрудникам реестра.


```python
employees_data = [
    (101, "Алиса", 1800, "Data"),
    (102, "Боб", 2200, "Backend"),
    (103, "Кира", 1600, "QA"),
    (104, "Данил", 2500, "Backend")
]


class Employee:
    def __init__(self, emp_id, name, salary, department):    # реализовать __init__
        self.emp_id = emp_id
        self.name = name
        self._salary = salary
        self.department = department
        
    @property     #  реализовать property salary (getter/setter с валидацией)
    def salary(self):
        return self._salary
    
    @salary.setter
    def salary(self, value):
        if value < 0:
            self._salary = 0
        else:
            self._salary = value

    def raise_salary(self, percent):       # увеличить salary на percent процентов
        self._salary = self._salary * (1 + percent/100)
    # новая salary = salary * (1 + percent / 100)
    def to_dict(self):    # вернуть dict: {"emp_id": ..., "name": ..., "salary": ..., "department": ...}
        return{
            "emp_id": self.emp_id,
            "name": self.name, 
            "salary": self.salary,
            "department": self.department
        }


class EmployeeRegistry:
    def __init__(self):
        self._employees = {}

    def add_employee(self, employee):
        self._employees[employee.emp_id] = employee     # добавить сотрудника в словарь по emp_id

    def remove_employee(self, emp_id):
        self._employees.pop(emp_id)    # удалить сотрудника по emp_id
        
    def find_by_department(self, department):
        return [emp for emp in self._employees.values() if emp.department == department]   # вернуть [emp for emp in self._employees.values() if emp.department == department]

    def total_payroll(self):
        return sum(emp.salary for emp in self._employees.values())    # вернуть сумму salary по всем self._employees.values()
        

registry = EmployeeRegistry()

for emp_id, name, salary, department in employees_data:
    employee = Employee(emp_id, name, salary, department)      # создать Employee и добавить в registry
    registry.add_employee(employee)

registry._employees[101].raise_salary(10)      #  повысить зарплату Алисе на 10%
registry._employees[103].raise_salary(5)        #  повысить зарплату Кире на 5%

print("Всего сотрудников:", len(registry._employees))
print("ФОТ:", registry.total_payroll())
print("Backend:", [e.name for e in registry.find_by_department("Backend")])
```
